####  String Methods and Slicing
###### Project Title: Feedback Analyzer
* Scenario:
* A content review team wants to analyse user feedback and comments. They want to check how
frequently certain words appear and extract parts of the feedback for reporting.
Tasks:
1. Ask the user to input a sentence (representing feedback).
2. Count how many words are in the sentence.
3. Check how many times the word “data” appears in the sentence, regardless of case.
4. Extract and print the first 5 and last 5 characters of the feedback separately.
5. Reverse the sentence and print it.

```python
# 1. Ask the user to input a sentence
feedback = input("Enter your feedback: ")

# 2. Count how many words are in the sentence
word_count = len(feedback.split())
print(f"Total words: {word_count}")

# 3. Check how many times the word “data” appears (case-insensitive)
data_count = feedback.lower().split().count("data")
# Alternative checking substring: feedback.lower().count("data")
print(f"The word 'data' appears: {data_count} times")

# 4. Extract and print the first 5 and last 5 characters
first_five = feedback[:5]
last_five = feedback[-5:] if len(feedback) >= 5 else feedback
print(f"First 5 characters: '{first_five}'")
print(f"Last 5 characters: '{last_five}'")

# 5. Reverse the sentence and print it
reversed_feedback = feedback[::-1]
print(f"Reversed sentence: {reversed_feedback}")
```

##### Project Title: Retail Inventory Manager
* Scenario:
You are managing an inventory system for a retail shop. Certain product categories are fixed,
but the actual products can be added or removed by the staff.
Tasks:
1. Create a tuple with fixed product categories like Electronics, Stationery, and Clothing.
2. Create a list of products that can be updated dynamically.
3. Ask the user to enter a new product and add it to the list.
4. Ask the user to input the index of a product to remove it from the list.
5. Display the final list of products and the product categories.


```python
# 1. Fixed product categories (Tuple)
categories = ("Electronics", "Stationery", "Clothing")

# 2. Dynamic product list (List)
products = ["Laptop", "Notebook", "T-Shirt"]

# 3. Add a new product
new_item = input("Enter a new product to add: ")
products.append(new_item)

# 4. Remove a product by index
print("\nCurrent product list:")
for idx, prod in enumerate(products):
    print(f"{idx}: {prod}")

try:
    rem_idx = int(input("Enter the index of the product to remove: "))
    removed = products.pop(rem_idx)
    print(f"Removed: {removed}")
except (ValueError, IndexError):
    print("Invalid index. No product removed.")

# 5. Display final lists
print("\n--- Final Shop Data ---")
print(f"Product Categories (Fixed): {categories}")
print(f"Current Inventory (Dynamic): {products}")
```

You have been assigned to build a contact management system where employees can store and

address.

```python
# 1. Create a 2D list of contacts [Name, Phone, Email]
contacts = [
    ["Alice Smith", "123-456", "alice@email.com"],
    ["Bob Jones", "987-654", "bob@email.com"]
]

# 2. Allow user to add a new contact
print("Add a new contact:")
name = input("Name: ")
phone = input("Phone: ")
email = input("Email: ")
contacts.append([name, phone, email])

# 3. Remove the last item
if contacts:
    removed_last = contacts.pop()
    print(f"Removed last contact: {removed_last}")

# 4. Get new items from users using input() and add them
print("\nAdd another contact setup:")
new_contact = [input("Enter Name: "), input("Enter Phone: "), input("Enter Email: ")]
contacts.append(new_contact)

# 5. Sort the list in ascending order based on Name (index 0)
contacts.sort(key=lambda x: x[0])

print("\n--- Sorted Contact Directory ---")
for c in contacts:
    print(f"Name: {c[0]} | Phone: {c[1]} | Email: {c[2]}")
```

####  Dictionary, Set, Frozenset
#### Project Title: Student Grades & Hobbies Tracker
* Scenario:
A student activity coordinator is collecting information about student grades and hobbies. They
want to ensure grades are updated properly and hobby lists are stored efficiently.
Tasks:
1. Create a dictionary with student names as keys and their grades as values.
2. Allow the user to add or update grades for at least 3 students.
3. Create a set of hobbies and allow the user to input at least 3 hobbies.
4. Convert the set of hobbies into a frozenset and display it.

```python
# 1. Create a dictionary with student names and grades
student_grades = {"John": 85, "Emma": 92}

# 2. Add or update grades for at least 3 students
for _ in range(3):
    s_name = input("Enter student name to add/update: ")
    s_grade = int(input(f"Enter grade for {s_name}: "))
    student_grades[s_name] = s_grade

# 3. Create a set of hobbies and input at least 3 hobbies
hobbies = set()
print("\nEnter 3 hobbies:")
for i in range(3):
    hobby = input(f"Hobby {i+1}: ")
    hobbies.add(hobby)

# 4. Convert the set into a frozenset and display it
frozen_hobbies = frozenset(hobbies)

print("\n--- Final Tracker Data ---")
print(f"Grades Dictionary: {student_grades}")
print(f"Hobbies Frozenset: {frozen_hobbies}")
```

#### Operators, Conditional Statements
#### Task:
* Given an integer, n, perform the following conditional actions:
*  If n is odd, print Weird
* If n is even and in the inclusive range of 2 to 5, print Not Weird
* If n is even and in the inclusive range of 6 to 20, print Weird
* If n is even and greater than 20, print Not Weird
Input Format :A single line containing a positive integer,n.
Constraints
* 1<=n <=100
Output Format: Print Weird if the number is weird. Otherwise, print Not Weird.

```python
n = int(input("Enter a positive integer (1-100): ").strip())

if n % 2 != 0:
    print("Weird")
else:
    if 2 <= n <= 5:
        print("Not Weird")
    elif 6 <= n <= 20:
        print("Weird")
    elif n > 20:
        print("Not Weird")
```

#### Project Title: Shopping Cart System
* Scenario:
* You are creating a simple billing system where customers can enter the prices of items they want to purchase. The program should keep accepting prices until the user decides to stop, then
apply appropriate discounts before generating the final bill.
* Tasks:
*  Continuously accept item prices using a while loop.
* Stop taking input when the user enters 0 (sentinel value).
* Keep track of the total number of items and the subtotal amount.
* Ensure no negative prices are allowed (ask again without counting as an item).
* Apply discounts based on the subtotal:
○ 20% for totals ≥ 1000
○ 10% for totals ≥ 500
○ 5% for totals ≥ 200
○ 0% for totals < 200
● Display a clear bill summary including subtotal, discount amount, and final total.


```python
subtotal = 0
total_items = 0

print("Enter item prices (Enter 0 to stop calculations):")
while True:
    try:
        price = float(input("Item price: "))
        if price == 0:
            break
        if price < 0:
            print("Negative prices not allowed. Try again.")
            continue

        subtotal += price
        total_items += 1
    except ValueError:
        print("Invalid number format. Try again.")

# Apply discounts based on total thresholds
if subtotal >= 1000:
    discount_pct = 0.20
elif subtotal >= 500:
    discount_pct = 0.10
elif subtotal >= 200:
    discount_pct = 0.05
else:
    discount_pct = 0.00

discount_amt = subtotal * discount_pct
final_total = subtotal - discount_amt

# Bill summary layout
print("\n--- BILL SUMMARY ---")
print(f"Total Items:     {total_items}")
print(f"Subtotal:        ${subtotal:.2f}")
print(f"Discount ({int(discount_pct*100)}%): ${discount_amt:.2f}")
print(f"Final Bill:      ${final_total:.2f}")
print("--------------------")
```

#### range, for loop, enumerate, list comprehension
#### Project Title: Employee Hours Report
* Scenario:
* You are analyzing student exam scores to generate performance reports.
* Tasks:
1. Create a list of 10 random exam scores ranging from 0 to 100 using list comprehension.
2. Display each score with its index using the enumerate function.
3. Find all scores greater than 75 and display them.
4. Calculate and display the average score.

```python
import random

# 1. Create a list of 10 random exam scores ranging from 0 to 100
scores = [random.randint(0, 100) for _ in range(10)]

# 2. Display each score with its index using enumerate
print("--- All Scores ---")
for index, score in enumerate(scores):
    print(f"Index {index}: Score = {score}")

# 3. Find all scores greater than 75
high_scores = [score for score in scores if score > 75]
print(f"\nScores greater than 75: {high_scores}")

# 4. Calculate and display average score
avg_score = sum(scores) / len(scores)
print(f"Average Exam Score: {avg_score:.2f}")
```

####  Built-in Functions and User-Defined Functions
###### Project Title: Simple Calculator App
###### Scenario:
* A company wants to provide users with a simple calculator to perform basic mathematical operations.
* Tasks:
1. Create functions for addition, subtraction, multiplication, and division.
2. Ask the user to input two numbers and the operation they want to perform.
3. Call the appropriate function and display the result.
4. Use built-in functions like abs(), round(), and pow() within your solution.
5. Handle division by zero gracefully and inform the user.


```python
# 1. Core operation functions
def add(a, b): return a + b
def subtract(a, b): return a - b
def multiply(a, b): return a * b
def divide(a, b):
    # 5. Handle division by zero gracefully
    if b == 0:
        return "Error: Division by zero is undefined."
    return a / b

# 2. Ask user for numeric and string inputs
try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    op = input("Enter operation (+, -, *, /): ").strip()

    # 3. Call functions & 4. Incorporate built-in functions inside execution logs
    if op == '+':
        res = add(num1, num2)
    elif op == '-':
        res = subtract(num1, num2)
    elif op == '*':
        res = multiply(num1, num2)
    elif op == '/':
        res = divide(num1, num2)
    else:
        res = "Invalid Operator!"

    print(f"\nResult: {res}")
    if isinstance(res, (int, float)):
        print(f"Absolute Value of Result: {abs(res)}")
        print(f"Rounded Value: {round(res, 2)}")
        print(f"Result raised to power of 2: {pow(res, 2)}")

except ValueError:
    print("Invalid numeric entry.")
```

###  Function Arguments, Local and Global Namespace, Introduction to OOP
#### Project Title: Bank Account Manager
###### Scenario:
* A bank wants to track account balances and transactions for each customer in an object-orientated way.
* Tasks:
1. Create a class BankAccount with account_holder and balance attributes.
2. Implement methods for deposit(), withdraw(), and display_balance().
3. Use both local and global variables appropriately, such as a global transaction counter.
4. Demonstrate calling methods using both positional and keyword arguments.
5. Prevent withdrawals that would result in a negative balance.


```python
# global transaction counter
transaction_counter = 0

# 1. Create the class
class BankAccount:
    def __init__(self, account_holder, balance=0.0):
        self.account_holder = account_holder
        self.balance = balance

    # 2. Methods for accounting transactions
    def deposit(self, amount):
        global transaction_counter
        self.balance += amount
        transaction_counter += 1
        print(f"Deposited ${amount:.2f}")

    def withdraw(self, amount):
        global transaction_counter
        # 5. Prevent withdrawals resulting in negative balance
        if amount > self.balance:
            print("Transaction Rejected: Insufficient funds!")
        else:
            self.balance -= amount
            transaction_counter += 1
            print(f"Withdrew ${amount:.2f}")

    def display_balance(self):
        print(f"Account Holder: {self.account_holder} | Current Balance: ${self.balance:.2f}")

# 4. Demonstrate calls with positional and keyword arguments
# Keyword argument instantiation
acct = BankAccount(account_holder="Sarah Connor", balance=500.0)

# Method calls
acct.display_balance()
acct.deposit(150.0)             # Positional
acct.withdraw(amount=200.0)      # Keyword
acct.withdraw(1000.0)            # Triggers failure check
acct.display_balance()

print(f"Total successful global transactions recorded: {transaction_counter}")
```

#### Project Title: Student Management System
#### Scenario:
* You are tasked with building a Student Management System that integrates multiple Python concepts to help staff manage students' data, performance, and reporting.
* Tasks:
1. Use dictionaries and lists to store and retrieve student information such as name, age,
and grades.
2. Allow users to add, delete, update, and search for students.
3. Track grades and categorize performance using functions and conditionals.
4. Use tuples or frozensets to store constant data like available classes.
5. Optionally implement file handling to save and load student data between program
runs.
6. Use exception handling to manage invalid inputs.
7. Implement sorting and filtering of students for reports.

```python
# 4. Available immutable classes setup
AVAILABLE_CLASSES = frozenset(["Math 101", "Science 102", "History 103"])

# 1. Storage backend dictionary mapping IDs or names to student profiles
students_db = {}

def add_student():
    name = input("Enter Student Name: ").strip()
    try:
        age = int(input("Enter Age: "))
        grade = float(input("Enter Grade Score (0-100): "))
        print(f"Available Tracks: {list(AVAILABLE_CLASSES)}")
        chosen_class = input("Assign standard class from list: ").strip()

        if chosen_class not in AVAILABLE_CLASSES:
            print("Warning: Class isn't an official track, assigning default.")
            chosen_class = "Unassigned"

        students_db[name] = {"age": age, "grade": grade, "class": chosen_class}
        print(f"Student {name} registered successfully!")
    except ValueError: # 6. Handle bad inputs
        print("Invalid entry details. Registration aborted.")

def delete_student():
    name = input("Enter student name to remove: ")
    if students_db.pop(name, None):
        print(f"Removed record for {name}")
    else:
        print("Student not found.")

def view_reports():
    # 7. Simple alphabetical structural reporting sort
    print("\n--- Sorted Student Performance Report ---")
    for name, data in sorted(students_db.items()):
        # 3. Categorize performance through conditional metrics
        status = "Passing" if data['grade'] >= 50 else "Failing"
        print(f"Name: {name} | Grade: {data['grade']} ({status}) | Class: {data['class']}")

# Small simulation runner for verification
add_student()
view_reports()
```
